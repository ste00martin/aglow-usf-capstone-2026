# Aglow Database Explorer
Sample data from each table in the database.

```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

load_dotenv()
engine = create_engine(os.getenv('DATABASE_URL'))
conn = engine.connect()
print('Connected!')
```

## Row Counts

```python
tables = ['profile', 'video', 'audio', 'feed', 'contact', 'notification', 'prompt', 'event', 'event_member', 'event_ticket_kind', 'file']
counts = {t: pd.read_sql(f'SELECT COUNT(*) as count FROM {t}', conn)['count'][0] for t in tables}
pd.DataFrame.from_dict(counts, orient='index', columns=['row_count']).sort_values('row_count', ascending=False)
```
